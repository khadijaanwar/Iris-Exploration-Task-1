# Task 1: Exploring and Visualizing a Simple Dataset

**Objective:**  
Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

**Dataset:**  
Iris Dataset (available via seaborn)

**Skills Covered:**
- Data loading and inspection using pandas
- Descriptive statistics and data exploration
- Basic plotting and visualization with seaborn and matplotlib


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset from seaborn
iris = sns.load_dataset('iris')
```

```python
# Print the shape of the dataset
print("Shape of the dataset:", iris.shape)

# Print the column names
print("Column names:", iris.columns.tolist())

# Display first few rows
print("First 5 rows:\n", iris.head())
```

```python
# General information about dataset
print("\nDataset Info:")
iris.info()

# Statistical summary
print("\nStatistical Summary:")
print(iris.describe())
```

```python
# Pairplot for feature relationships
sns.pairplot(iris, hue='species')
plt.suptitle("Pairwise Scatter Plots", y=1.02)
plt.show()
```

```python
# Histogram for each feature
iris.hist(edgecolor='black', figsize=(10, 6))
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
# Box plot for each feature grouped by species
plt.figure(figsize=(12, 6))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(1, 4, i)
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f"{column.capitalize()} Boxplot")
plt.tight_layout()
plt.show()
```
